# Praktik Spark Lanjutan
**Latihan 1**: Memanupulasi Data dengan PySpark



```python
from pyspark.sql import SparkSession

# Memulai Spark Session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

#Membuat DataFrame Sederhana
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

#Menampilkan DataFrame
df.show()

#Filtering Data
df_filtered = df.filter(df['usia'] > 30)
df_filtered.show()

# menghitung rata-rata usia
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

#mengurutkan data berdasarkan usia
df_sorted =  df.orderBy("Usia", asending=False)
df_sorted.show()
```

**Tugas 2**: Lakukan filter, penghitungan rata-rata, dan pengurutan data menggunakan PySpark.

```python
from pyspark.sql import SparkSession

# Memulai Spark Session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

#Membuat DataFrame Sederhana
data = [("Ali", 34, "penjahit", "renang", "laki-laki"),
 ("Budi", 23, "Programmer", "gitar", "laki-laki"),
 ("Citra", 29, "data analyst", "baca buku", "perempuan"),
 ("Dina", 45, "web dev", "coding", "perempuan")]
columns = ["Nama", "Usia", "pekerjaan", "hobi", "gender"]
df = spark.createDataFrame(data, columns)

#Menampilkan DataFrame
print("DataFrame")
df.show()

#Filtering Data
print("Data difilter berdasarkan gender (laki-laki)")
df_filtered = df.filter(df['gender'] == "laki-laki")
df_filtered.show()

# menghitung rata-rata usia
print("rata-rata usia")
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

#mengurutkan data berdasarkan usia
print("Data diurutkan berdasarkan usia")
df_sorted =  df.orderBy("Usia", ascending=False)
df_sorted.show()
```

**Tugas 5**: Gunakan metode ini untuk menggabungkan data yang Anda buat di PySpark dengan data dari Pandas, kemudian lakukan analisis sederhana seperti menghitung rata-rata usia.


```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import avg
import pandas as pd

# Memulai Spark Session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# DataFrame PySpark
data = [
    ("Ali", 34, "penjahit", "renang", "laki-laki"),
    ("Budi", 23, "Programmer", "gitar", "laki-laki"),
    ("Citra", 29, "data analyst", "baca buku", "perempuan"),
    ("Dina", 45, "web dev", "coding", "perempuan")
]
columns = ["Nama", "Usia", "Pekerjaan", "Hobi", "Gender"]

df_spark = spark.createDataFrame(data, columns)

# DataFrame Pandas
data_pandas = {
    "Nama": ["Ali", "Budi", "Citra", "Dina"],
    "Kota": ["Jakarta", "Bandung", "Surabaya", "Medan"]
}
df_pandas = pd.DataFrame(data_pandas)

# Mengonversi DataFrame Pandas ke PySpark
df_spark_from_pandas = spark.createDataFrame(df_pandas)

# Menggabungkan (Join) DataFrame Spark
df_joined = df_spark.join(df_spark_from_pandas, on="Nama")

print("Hasil Join antara PySpark dan Pandas:")
df_joined.show()

# Analisis sederhana: rata-rata usia
print("Rata-rata Usia:")
df_joined.groupBy().agg(avg("Usia")).show()
```

# Konversi Data antara PySpark dan Pandas

**Tugas 6**: Gabungkan data dari PySpark dan Pandas, lalu lakukan operasi statistik seperti menghitung nilai maksimum usia.

```python
from pyspark.sql import SparkSession
import pandas as pd

# DataFrame PySpark
spark = SparkSession.builder.appName("GabungSparkPandas").getOrCreate()

data_spark = [
    ("Ali", 34, "penjahit", "renang", "laki-laki"),
    ("Budi", 23, "Programmer", "gitar", "laki-laki"),
    ("Citra", 29, "data analyst", "baca buku", "perempuan"),
    ("Dina", 45, "web dev", "coding", "perempuan")
]
columns_spark = ["Nama", "Usia", "Pekerjaan", "Hobi", "Gender"]
df_spark = spark.createDataFrame(data_spark, columns_spark)

# DataFrame Pandas
data_pandas = {
    "Nama": ["Ali", "Budi", "Citra", "Dina"],
    "Kota": ["Jakarta", "Bandung", "Surabaya", "Medan"]
}
df_pandas = pd.DataFrame(data_pandas)

# Konversi PySpark DataFrame ke Pandas
df_spark_to_pandas = df_spark.toPandas()

# Gabungkan kedua DataFrame berdasarkan kolom "Nama"
df_joined = pd.merge(df_spark_to_pandas, df_pandas, on="Nama")
print("Hasil Gabungan:")
print(df_joined)

# Analisis sederhana: nilai maksimum usia
max_usia  = df_joined["Usia"].max()
nama_max = df_joined[df_joined["Usia"] == max_usia]["Nama"].values[0]
print(f"\nNilai maksimum usia adalah {max_usia} yang dimiliki oleh {nama_max}")
```
